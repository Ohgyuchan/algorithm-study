```python
# Recursive function
def recur_sum(x):
    if x < 1:
        return x
    return x + recur_sum(x-1)
print(recur_sum(10))
```

```python
def sum(n):
    if n < 1:
        return n
    return n + sum(n-1)
sum(10)
```

```python
def is_odd(n):
    if (n%2) == 0 :
        return "Even"
    else :
        return "Odd"
is_odd(13)
```
